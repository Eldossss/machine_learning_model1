```python
import pandas as pd
```

```python
houses_file_path = 'data.csv'
```

```python
houses_data  = pd.read_csv(houses_file_path)
```

```python
houses_data.describe()
```

# ** Это у нас для того чтобы дропнуть все not available значения

```python
houses_data = houses_data.dropna(axis = 0)
```

```python
houses_data.describe()
```

# ** Мы берем как прайс значение Y

```python
y = houses_data.price
```

```python
y.describe()
```

# ** Мы берем все значение по каким параметрам мы будем предиктовать прайс дома

```python
houses_features =  ['bedrooms','bathrooms','floors']
```

```python
x = houses_data[houses_features]
```

```python
    x.describe()
```

# ** Мы через scikit-learn библиотеку берем десижн три дерево чтобы определить цену домов 

```python
from sklearn.tree import DecisionTreeRegressor 
```

```python
house_model = DecisionTreeRegressor(random_state = 1)
```

```python
house_model.fit(x,y)
```

```python
print("Making predictions for the following 5 houses")
```

```python
print(x.head())
```

```python
print("The predictions are")
```

```python
print(house_model.predict(x.head()))
```
